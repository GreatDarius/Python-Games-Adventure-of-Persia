```python
life = 3
import time
import random
delay = (random.randint(0,9))/10

def win():
    time.sleep(delay)
    print("You have succesfully finished your quest")


def failed():
    global life
    life = life-1
    if life == 0:
        print("I'm AFRAID my lord there is NO MORE LIVES sire!!")
    else: 
        time.sleep(delay)
        print("You made too many mistakes! And Died")


def treasury():
    place_creator([True,False],"The Treasury ",True,["win","failed"])

def main_gate():
    place_creator([True],"The Main Gate ",False,["treasury"])


def right_staires():
    place_creator([True],"The Right Side of the Stairecase ",False,["main_gate"])


def left_staires():
    place_creator([False],"The left Side of the Stairecase ",False,["failed"])


def staires():
    place_creator([True,False],"The Staires ",False,["left_staires","right_staires"])


def abandoned_building():
    place_creator([True],"The Abandoned Building ",False,["main_road"])


def forest():
    place_creator([False],"The Forest ",False,["failed"])


def river():
    place_creator([False],"The River ",False,["failed"])


def main_road():
    place_creator([False,False,True,True], "The Main Road", False, ["river","forest","abandoned_building","staires"])

for i in range(life):
    if life >0:

        def place_creator(place_list, place_name, is_end, next_place):
            print(f"You are now in {place_name}.")
            time.sleep(delay)
            place_count = len(place_list)
            choices = []

            input_message= "Choose a path:("

            place_message = ""
            for count in range(place_count):
                choices.append(f"place {count + 1}")
                if count + 1 < place_count:
                    place_message += f"place {count + 1},"
                else:
                    place_message += f"place {count + 1}):"
            input_message += place_message
            time.sleep(delay)

            while True:
                command = input(input_message)

                if command in choices:
                    if place_list[choices.index(command)]:
                        if is_end:
                            time.sleep(delay)
                            print("Well done my LORD!")
                            win()
                            break
                        else:
                            time.sleep(delay)
                            print("Move on my Lord time is limited")
                            time.sleep(delay)
                            eval(next_place[choices.index(command)] + "()")
                            break
                    else:
                        failed()
                        break
                else:
                    time.sleep(delay)
                    print(f"I'm afraid my LORD you must choose from these path: ({place_message})")

        place_creator([False,False,True,True], "The Main Road", False, ["river","forest","abandoned_building","staires"])
  




```
